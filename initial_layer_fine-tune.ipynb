```python
from ultralytics import YOLO
```

```python
# Load the pre-trained YOLO model
model = YOLO("/content/drive/MyDrive/YOLO/run/runs/detect/train/weights/best.pt")
```

```python
print(model)
```

Train only some initial few layer and keep other freeze

```python
for name, param in model.model.named_parameters():

    if any(f"model.{i}." in name for i in range(4)):
        param.requires_grad = True
    else:
        param.requires_grad = False

    # Print for verification (optional)
    print(f"{name}: {'Trainable' if param.requires_grad else 'Frozen'}")
```

```python
model.train(
    data='/content/drive/MyDrive/YOLO/Adverse_data_30/data_adv.yaml',
    epochs=50,
    imgsz=640,
    lr0=0.0001

)
```
